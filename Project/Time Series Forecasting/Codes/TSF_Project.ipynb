# COMBINED DATASET

```python
from functools import reduce
stocks = reduce(lambda x,y: pd.merge(x,y, on='Date', how='outer'), [LSE,SSE,Nikkei225,Dow_Jones, Nasdaq, Nifty50])
```

```python
stocks.info()
```

```python
st = stocks.copy()
st
```

```python
st["LSE"] = pd.to_numeric(st.LSE, errors='coerce')
st["SSE"] = pd.to_numeric(st.SSE, errors='coerce')
st["Nasdaq"] = pd.to_numeric(st.Nasdaq, errors='coerce')
st["Dow Jones"] = pd.to_numeric(st["Dow Jones"], errors='coerce')
st["Nifty50"] = pd.to_numeric(st.Nifty50, errors='coerce')
st["Nikkei225"] = pd.to_numeric(st.Nikkei225, errors='coerce')
```

```python
st_nonmiss = st.dropna()
med_LSE = np.median(st_nonmiss['LSE'])
med_SSE = np.median(st_nonmiss['SSE'])
med_NAS = np.median(st_nonmiss['Nasdaq'])
med_NIK = np.median(st_nonmiss['Nikkei225'])
med_DOW = np.median(st_nonmiss['Dow Jones'])
med_NIF = np.median(st_nonmiss['Nifty50'])
```

```python
med_LSE
```

```python
st['LSE'].fillna(value =  med_LSE, inplace = True)
```

```python
st['SSE'].fillna(value =  med_SSE, inplace = True)
st['Nasdaq'].fillna(value =  med_NAS, inplace = True)
st['Nikkei225'].fillna(value =  med_NIK, inplace = True)
st['Dow Jones'].fillna(value =  med_DOW, inplace = True)
st['Nifty50'].fillna(value =  med_NIF, inplace = True)
```

```python
st['SSE'] = np.log(st['SSE'] + 1 - min(st['SSE']))
st['Nasdaq'] = np.log(st['Nasdaq'] + 1 - min(st['Nasdaq']))
st['Nikkei225'] = st['Nikkei225'].apply(lambda x : np.log(x) if x>0 else (-np.log(x)))
st['Dow Jones'] = st['Dow Jones'].apply(lambda x : np.log(x) if x>0 else (-np.log(x)))
st['Nifty50'] = st['Nifty50'].apply(lambda x : np.log(x) if x>0 else (-np.log(x)))
st['LSE'] = st['LSE'].apply(lambda x : np.log(x) if x>0 else (-np.log(x)))
```

```python
cols = ['Nasdaq', 'SSE', 'LSE', 'Nikkei225', 'Dow Jones', 'Nifty50']
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer()
print(pt.fit(st[cols]))
Date = st.drop(cols, axis = 1)
sto = pd.DataFrame(pt.transform(st[cols]),columns= cols)
st_new = pd.concat([Date, sto], axis = 1)
```
