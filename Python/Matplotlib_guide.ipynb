```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

LINE CHART

```python
plt.figure(figsize=[3,3],dpi=140)
x = np.arange(10)
y = x+2
x1=np.arange(5)
plt.plot(x,y,label='x+2',color='green',linewidth=3,linestyle='-',marker='.',markersize=10,markeredgecolor='red')
plt.plot(x1[:3],x1[:3]**2,'r*-',label='x^2')
plt.plot(x1[2:],x1[2:]**2,'r*--')
plt.title('Linear Graph', fontdict={'fontname':'Comic Sans MS'},fontsize=20)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xticks([0,2,4,6,8,10])
plt.yticks([0,5,10,15,20])
plt.legend()
#plt.savefig('mygraph.png',dpi=300)
plt.show()
```

BAR CHART

```python
labels = ['A','B','C','D']
values = [2,4,6,8]

bars = plt.bar(labels,values)
patterns = ['/','o','*','x']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('*')
#bars[3].set_hatch('x')
plt.show()
```
